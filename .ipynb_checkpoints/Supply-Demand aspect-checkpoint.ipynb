```python
import pandas as pd
```

```python
supply = pd.read_csv(r"C:\Users\vishal\OneDrive\Desktop\US House Price Project\Data\supply.csv")
demand = pd.read_csv(r"C:\Users\vishal\OneDrive\Desktop\US House Price Project\Data\demand.csv")
```

```python
start_date1 = '01-01-2003'
end_date1 = '01-10-2022'

supply_filtered = supply[(supply['DATE'] >= start_date1) & (supply['DATE'] <= end_date1)]
demand_filtered = demand[(demand['DATE'] >= start_date1) & (demand['DATE'] <= end_date1)]

# Convert the 'DATE' column to datetime format and back to the desired format 'yyyy-mm-dd'
supply_filtered['DATE'] = pd.to_datetime(supply_filtered['DATE'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')
demand_filtered['DATE'] = pd.to_datetime(demand_filtered['DATE'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')

# Print the updated DataFrames
```
